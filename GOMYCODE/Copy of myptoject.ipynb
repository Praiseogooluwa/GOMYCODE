SHOPPING LIST BY OG_GRAPHIX

```python
# Initialize an empty list to store shopping items
shopping_list = []

# Main program
while True:
    # Display the menu options
    print("\nShopping List Menu:")
    print("1. Add item")
    print("2. Remove item")
    print("3. View items")
    print("4. Quit")

    # Get user input for menu selection
    choice = input("Enter your choice (1-4): ")

    # Perform actions based on user's choice
    if choice == '1':  # Add item
        if len(shopping_list) < 5:  # Limit the number of items using range() function
            item = input("Enter the item to add: ")
            shopping_list.append(item)
            print(f"{item} added to the shopping list.")
        else:
            print("Maximum number of items reached. Cannot add more.")
    elif choice == '2':  # Remove item
        item_to_remove = input("Enter the item to remove: ")
        if item_to_remove in shopping_list:
            shopping_list.remove(item_to_remove)
            print(f"{item_to_remove} removed from the shopping list.")
        else:
            print(f"{item_to_remove} is not in the shopping list.")
    elif choice == '3':  # View items
        print("Items in the shopping list:")
        for item in shopping_list:
            print("-", item)
    elif choice == '4':  # Quit
        print("Exiting the shopping list program. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 4.")
```
